```python
%matplotlib inline
import random

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.style.use('ggplot')
random.seed()
```

```python
train = pd.read_csv('Features_Variant_1.csv', sep=',')
train.head()
```

```python
# Нормировка и сохранение коэффициентов
norm_coeffs = {}
for column in train.columns[:-1]:
    mean = np.mean(train[column])
    std = np.std(train[column])
    norm_coeffs[column] = {'mean': mean, 'std': std}
    train[column] = (train[column] - mean) / std
train.head()
```

```python
tests = [pd.read_csv(f'Test_Case_{i}.csv', sep=',', header=None, names=train.columns) for i in range(1, 11)]
test = pd.concat(tests, ignore_index=True)
for column in train.columns[:-1]:
    test[column] = (test[column] - norm_coeffs[column]['mean']) / norm_coeffs[column]['std']
print(test.shape)
test.head()
```

```python
# Разбиение на X и y
# bias will be first
X_train = train.drop(columns=['output'])
tmp = X_train.columns.tolist()
X_train['bias'] = 1
X_train = X_train[['bias'] + tmp]
y_train = train['output']

X_test = test.drop(columns=['output'])
tmp = X_test.columns.tolist()
X_test['bias'] = 1
X_test = X_test[['bias'] + tmp]
y_test = test['output']

X_np = X_train.to_numpy()
y_np = y_train.to_numpy()

X_test_np = X_test.to_numpy()
y_test_np = y_test.to_numpy()

X_train.head()
```

```python
def shuffled(X: np.ndarray, y: np.ndarray):
    zipped = np.hstack((X, y.reshape(-1, 1)))
    np.random.shuffle(zipped)
    shuffled_X, shuffled_y = np.hsplit(zipped, np.array([-1]))
    return shuffled_X, shuffled_y.reshape(-1)


def mae(fact: np.ndarray, pred: np.ndarray):
    return np.mean((fact - pred) ** 2)


def mae_(fact: np.ndarray, pred: np.ndarray):
    return mae(fact, pred) / 2


def rmse(fact: np.ndarray, pred: np.ndarray):
    return np.sqrt(np.mean((fact - pred) ** 2))


def r2(fact: np.ndarray, pred: np.ndarray):
    return 1 - mae(fact, pred) / mae(fact, np.mean(fact))


def loss_and_grad(X: np.ndarray, y: np.ndarray, thetas: np.ndarray,
                  reg: float):
    n, m = X.shape
    preds = X @ thetas
    X_tr = X.transpose()
    
    main_grad = 1 / n * X_tr @ (preds - y)
    reg_grad = reg / (m * n) * thetas
    # По bias не пускают регуляризацию
    reg_grad[0] = 0
    result_grad = main_grad + reg_grad
    return mae_(y, preds), result_grad


def grad_descent(X: np.ndarray, y: np.ndarray, lr: float, reg: float,
                 stopping_thresh: float, max_iters: int = 100):
    n, m = X.shape
    assert y.shape == (n,)
    thetas = np.random.randn(m) * np.sqrt(1 / m)
    last_loss = -100
    for _ in range(max_iters):
        loss, grad = loss_and_grad(X, y, thetas, reg)
        if np.abs(loss - last_loss) < stopping_thresh:
            break
        thetas = thetas - lr * grad
        last_loss = loss

    return thetas


def get_kfold_errors(X: np.ndarray, y: np.ndarray, lr: float, reg: float,
                     stopping_thresh: float, n_folds: int = 5):
    assert len(y.shape) == 1, X.shape[0] == y.shape[0]
    folds_X = np.array_split(X, n_folds)
    folds_y = np.array_split(y, n_folds)

    val_errors = []
    for i in range(n_folds):
        current_train_X = np.vstack([folds_X[j]
                                     for j in range(n_folds) if j != i])
        current_train_y = np.hstack([folds_y[j]
                                     for j in range(n_folds) if j != i])
        current_val_X = folds_X[i]
        current_val_y = folds_y[i]
        thetas = grad_descent(current_train_X, current_train_y, lr, reg,
                              stopping_thresh)
        current_error = rmse(current_val_y, current_val_X @ thetas)
        val_errors.append(current_error)

    return val_errors
```

```python
tested_params = []
```

## Тренировка и валидация с руками выставленными параметрами

```python
hyperparams = {
    'lr': 0.1,
    'reg': 0.1,
    'stopping_thresh': 1
}

X_np, y_np = shuffled(X_np, y_np)
errors = get_kfold_errors(X_np, y_np, **hyperparams, n_folds=5)
print('Fold errors (RMSE): ', end='')
for err in errors:
    print(f'{err:.3f}  ', end='')
print(f'\nMean: {np.mean(errors):.3f}\n')

thetas = grad_descent(X_np, y_np, **hyperparams)
train_pred = X_np @ thetas
test_pred = X_test_np @ thetas
print(f'\033[4m      |  RMSE  |  R^2  \033[0m')
print(f'\033[4mTrain | {rmse(y_np, train_pred):6.2f} | {r2(y_np, train_pred):5.3f} \033[0m')
print(f'Test  | {rmse(y_test_np, test_pred):6.2f} | {r2(y_test_np, test_pred):5.3f} ')
```

```python
def get_random_hyperparams():
    hyperparams = {}
    min_lr = 0.001
    max_lr = 0.1
    hyperparams['lr'] = random.uniform(min_lr ** 0.5, max_lr ** 0.5) ** 2
    min_reg = 0.001
    max_reg = 1
    hyperparams['reg'] = random.uniform(min_reg ** 0.5, max_reg ** 0.5) ** 2
    min_stopping_thresh = 0.1
    max_stopping_thresh = 100
    hyperparams['stopping_thresh'] = random.uniform(
        min_stopping_thresh ** 0.5, max_stopping_thresh ** 0.5) ** 2
    
    return hyperparams
```

```python
for i in range(100):
    print(i, end=' ')
    hyperparams = get_random_hyperparams()
    errors = get_kfold_errors(X_np, y_np, **hyperparams, n_folds=5)
    error = np.mean(errors)
    tested_params.append((hyperparams, error))
```

```python
best_params = sorted(tested_params, key=lambda x: x[1])
best_params[:5]
```

## Тренировка и валидация с параметрами, найденными случайным поиском

```python
hyperparams = best_params[0][0]
thetas = grad_descent(X_np, y_np, **hyperparams)
train_pred = X_np @ thetas
test_pred = X_test_np @ thetas
print(f'\033[4m      |  RMSE  |  R^2  \033[0m')
print(f'\033[4mTrain | {rmse(y_np, train_pred):6.2f} | {r2(y_np, train_pred):5.3f} \033[0m')
print(f'Test  | {rmse(y_test_np, test_pred):6.2f} | {r2(y_test_np, test_pred):5.3f} ')
```

## Картина не сильно поменялась...

```python
from scipy import stats


def regression_coef(preds, thetas, X, y):
    coef = pd.DataFrame(list(zip(X.columns.tolist(), thetas.tolist())),
                        columns=['predictor', 'coef'])
    X1 = X.to_numpy()
    b = thetas
    MSE = np.sum((preds - y) ** 2, axis=0) / float(X.shape[0] - X.shape[1])
    var_b = MSE * (np.linalg.pinv(np.dot(X1.T, X1)).diagonal())
    sd_b = np.sqrt(var_b)
    t = b / sd_b
    coef['pvalue'] = [2 * (1 - stats.t.cdf(np.abs(i), (len(X1) - 1))) for i in t]
    return coef
```

```python
# Да, можно было просто посмотреть на модуль самих коэффициентов, но мы не ищем простых путей
pvalue_table = regression_coef(train_pred, thetas, X_train, y_np)
pvalue_table
```

### Возьмём только значимые фичи

```python
good_features = pvalue_table[pvalue_table['pvalue'] < 0.05]['predictor']
good_features
```

```python
X_train_good = X_train[good_features.to_list()].to_numpy()
y_train_good = y_train.to_numpy()

hyperparams = best_params[0][0]
thetas = grad_descent(X_train_good, y_train_good, **hyperparams)
train_pred = X_train_good @ thetas
X_test_np = X_test[good_features.to_list()]
test_pred = X_test_np @ thetas
print(f'\033[4m      |  RMSE  |  R^2  \033[0m')
print(f'\033[4mTrain | {rmse(y_train_good, train_pred):6.2f} | {r2(y_train_good, train_pred):5.3f} \033[0m')
print(f'Test  | {rmse(y_test_np, test_pred):6.2f} | {r2(y_test_np, test_pred):5.3f} ')
```
